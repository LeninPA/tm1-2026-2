## Ejemplo 7.2.2. Enteros positivos menores a mil que no son divisibles ni por dos ni tres ni cinco

Para obtener las combinaciones puedes usar `combinations` para crearlas y `comb` para contarlas

```python
N, primos = 1000, (2,3,5)
def mult(p: int) -> int: return N // p
total = 0
# Fórmula de inclusión-exclusión
for r in range(1, len(primos) + 1):
  for subcjto in combinations(primos, r):
    p = prod(subcjto)
    total += (-1) ** (r + 1) * mult(p)
print(N - total)
```

## Ejemplo 7.2.3. Locuras y números de Catalan

**Def.** Una *locura* es una permutación $\sigma\in S_n$ tal que para toda $i\in\mathbb{N}^+$ con $1\leqslant i\leqslant n$, $\sigma(i)\ne i$

```python
def random_perm(n: int) -> list[int]:
  p = list(range(n))
  shuffle(p)
  return p

def q_es_locura(p: list[int]) -> bool:
  # ¿Hay alguna manera más eficiente de escribirlo?
  return all(p[i] != i for i in range(len(p)))

n, intentos = 9, 50_000
total = sum(q_es_locura(random_perm(n)) for _ in range(intentos))

print(f"Encontramos {total} locuras")
print(f"La probabilidad de encontrarlas es {(total / intentos) * 100:.2f}%")
```
